## 函式學習

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 5, 50)
```

```python
y = 1.2*x + 0.8 
```

```python
plt.scatter(x, y)
```

### 加上 noise

```python
y = 1.2*x + 0.8 + 0.5*np.random.randn(50)
```

```python
plt.scatter(x, y, c=y, cmap="coolwarm")
plt.plot(x, 1.2*x + 0.8, 'r')
```

### 做線性回歸 by sci-kit learn

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

### 將 x 改成 nx1 array

```python
X = x.reshape(-1, 1)
print(X.shape)
print (y.shape)
```

```python
reg.fit(X, y)
```

```python
test = np.linspace(6, 10, 50)
test = test.reshape(-1, 1)
print (test.shape)
```

```python
Y = reg.predict(X)
```

```python
print (x.shape)
print (Y.shape)
```

```python
plt.scatter(x, y)
plt.plot(x, Y, "r")
plt.plot(x, 1.2*x+0.8, 'g')
```

```python
predY = reg.predict(test)
print (predY.shape)
```

```python
plt.scatter(test, predY, c='g')
plt.plot(x, 1.2*x+0.8, 'm')
```

### 取資料分 test train

```python
x = np.linspace(0, 5, 200)
y = 1.2*x + 0.8 + 0.5*np.random.randn(200)
```

```python
plt.scatter(x, y, c='m')
```


### 將資料分成 70:30

```python
from sklearn.model_selection import train_test_split
```

```python
X = x.reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 87)
```

```python
plt.scatter(x_train, y_train, c='m')
plt.scatter(x_test, y_test, c='b')
```

### 接下來可以開始 training 了!!

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
plt.scatter(x_test, y_test, c='r', label = "data")
plt.plot(x_test, y_pred, 'b', label = "predicted")
plt.plot(x, 1.2*x+0.8, 'g', label = "label")
plt.legend();
```
