## GroupBy

```python
ufo = pd.read_csv("https://bit.ly/uforeports")
```

```python
ufo.head()
```

```python
ufo_state = ufo.groupby("State")
```

groupby會將 Nan值直接自動忽略，因此每個row中的值會不一樣。

```python
ufo_state = ufo_state.count()
```

```python
ufo_state.sort_values(by="Time", ascending = False, inplace = True)
```

```python
ufo_state[:10].Time.plot(kind="bar")
```

```python
ufo.groupby(["State", "City"]).count()
```

```python
ufo.groupby(["State", "City"]).count().reset_index()
```

## datetime

```python
ufo = pd.read_csv("https://bit.ly/uforeports")
ufo.head()
```

```python
ufo["datetime"] = pd.to_datetime(ufo.Time)
```

```python
ufo.head()
```

```python
ufo["month"] = ufo.datetime.dt.month
```

```python
ufo.head()
```

```python
ufo["weekday"] = ufo.datetime.dt.weekday
```

```python
ufo.head()
```

## 星期幾比較容易看到幽浮??

```python
ufo_weekdays = ufo.groupby("weekday").count()
```

```python
ufo_weekdays.sort_values(by="Time", ascending=False, inplace=True)
```

```python
ufo_weekdays.Time.plot(kind="bar")
```
