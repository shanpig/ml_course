```python
dataframe=pd.read_csv("http://stats.moe.gov.tw/files/detail/107/107_student.csv")
```

## 標題被放到第一列，修改檔案使其正確

```python
colname = dataframe.loc[1].values
```

```python
colname
dataframe = dataframe[2:]
dataframe.columns = colname
dataframe
```

```python
colname
```

```python
undergraduate = dataframe[(dataframe["日間∕進修別"] == 'D 日') & (dataframe["等級別"] == "B 學士")]
undergraduate
```

```python
df2 = undergraduate.loc[:, "學校名稱" : "女生計"].copy()
df2
```

#### 資料裡含有逗號，需要移除

```python
df2["男生計"]/df2["女生計"] # will cause error
```

```python
import locale
# atof  變成浮點數
```

```python
locale.setlocale(locale.LC_NUMERIC, "")
```

### 問：為何不把資料用迴圈改？查!!!

```python
df2['男生計'] = df2['男生計'].apply(locale.atof)
df2['女生計'] = df2['女生計'].apply(locale.atof)
```

```python
df2["男女比例"] = df2["男生計"] / df2["女生計"]
df2.head()
```

```python
df3 = df2.sort_values(by=["男女比例"], ascending = False)
```

```python
df3[:20].plot.bar(x="學校名稱", y="男女比例")
```
